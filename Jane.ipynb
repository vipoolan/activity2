Note: color is a factor variable. So, create dummy variables with 0/1 values for each category value (One-hot encoding).

```python
janedummy = pd.get_dummies(data=jane, drop_first=True)
print(janedummy.head())
```

# Include both x and color as predictors

```python
model1 = ols('y ~ x + color', data=jane).fit()
print(model1.summary())
```

Note: intercept: mean for color=blue and x=0, and coefficient for colorgreen: difference in means for color=green and color=blue.

# Is color significant?

```python
model2 = ols('y ~ x', data=jane).fit()
print(sm.stats.anova_lm(model2,model1))
```

# Is there interaction between x and color?

```python
model3 = ols('y ~ x + color + x:color', data=jane).fit()
print(model3.summary())
```

```python
print(sm.stats.anova_lm(model1,model3))
```

# Confidence interval and prediction interval

```python
newdata = pd.DataFrame({"x":[2],"color":["blue"]})
pred = model1.predict(newdata)
print(pred)
```
